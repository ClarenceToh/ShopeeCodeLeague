```
# import specific libraries
# load CSVs
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```

data = pd.read_csv('order_brush_order.csv.zip')
```

```
data['datetime'] =  pd.to_datetime(data['event_time'])#, format='%Y%m%d:%H:%M:%S.%f')
```

```
removed = data[data.groupby('shopid').shopid.transform(len) > 2]
removed
```

```
unique_shops = list(data.shopid.unique())
unique_shops
```

```
sort = removed.sort_values(['shopid','userid'])#.sort_values('userid')

# byhr = sort.groupby(sort['datetime'].map(lambda x: x.hour))
sort.head(30)
```

```
def most(id):
    test = sort.loc[sort['shopid'] == id]
    test.sort_values('datetime')
    test.index = list(range(len(test)))
    num = int(len(test) - test.groupby('shopid')['userid'].nunique())
    if num > 1:
        return id
most(10084)
```

```
for_test=[]
for i in list(sort.shopid.unique()):
    for_test.append(most(i))
```

```
for_test = [x for x in for_test if x is not None]
```

```
len(for_test)
```

```
latest = sort[sort['shopid'].isin(for_test)]
latest
```

```
from datetime import timedelta
import more_itertools as mit
def fraud(id):
    ind=[]
    shops = []
    users = []
    tryy = latest[latest['shopid'] == id]
    tryy.sort_values(by='datetime')
    tryy.index = list(range(len(tryy)))
    same=[]
    for i in range(len(tryy)):
        try:
            if tryy['datetime'][i]-tryy['datetime'][i-1] < timedelta(hours=1) and tryy['datetime'][i]-tryy['datetime'][i-1] > timedelta(seconds=0):
                same.append(i-1)
                same.append(i)
        except:
            continue
    same = list( dict.fromkeys(same) )
    final=[list(group) for group in mit.consecutive_groups(same)]
    for i in final:
        if len(i)>=3:
            ind.append(i)
    finall = [item for items in ind for item in items]
    finall = list( dict.fromkeys(finall) )
    for y in finall:
        users.append(str(tryy['userid'][y]))
        return  id, '&'.join(users)
fraud(10100)
```

```
list_fraud=[]
for i in for_test:
    list_fraud.append(fraud(i))
```

```
list_fraud=[x for x in list_fraud if x is not None]
list_fraud
```
